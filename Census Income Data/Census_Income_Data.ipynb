```python
column_name = {
    'AAGE': 'Age',
    'ACLSWKR': 'Class_of_Worker',
    'ADTINK': 'Industry_Code',
    'ADTOCC': 'Occupation_Code',
    'AHGA': 'Education',
    'AHSCOL': 'Enrolled_in_Edu_Inst_Last_Week',
    'AMARITL': 'Marital_Status',
    'AMJIND': 'Industry',
    'AMJOCC': 'Occupation',
    'ARACE': 'Race',
    'PEMNTVTY': 'Country_of_Birth_Self',
    'PENATVTY': 'Country_of_Birth_Father',
    'PRCITSHP': 'Country_of_Birth_Mother',
    'SEOTR': 'Citizenship',
    'VETQVA': 'Own_Business_or_Self_Employed',
    'VETYN': 'Veteran_Benefits',
    'WKSWORK': 'Weeks_Worked_in_Year',
    'year': 'Year',
    'income_bracket': 'Income_Bracket'
}
```

```python
df= dataset.rename(columns=column_name)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Income_Bracket', data=df)
plt.title('Income_Bracket vs Age')
plt.xlabel('Age')
plt.ylabel('Income_Bracket')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Race', y='Income_Bracket', data=df)
plt.title('Income_Bracket vs Race')
plt.xlabel('Race')
plt.ylabel('Income_Bracket')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education', y='Income_Bracket', data=df)
plt.title('Income_Bracket vs Education')
plt.xlabel('Education')
plt.ylabel('Income_Bracket')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Occupation_Code', y='Income_Bracket', data=df)
plt.title('Income_Bracket vs Occupation_Code')
plt.xlabel('Occupation_Code')
plt.ylabel('Income_Bracket')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Marital_Status', y='Income_Bracket', data=df)
plt.title('Income_Bracket vs Marital_Status')
plt.xlabel('Marital_Status')
plt.ylabel('Income_Bracket')
plt.xticks(rotation=45)
plt.show()
```
